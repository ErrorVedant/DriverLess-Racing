```python
X_train, X_test, y_train, y_test=train_test_split(X,y,random_state=104,test_size=0.3,shuffle=True)
```

```python
svr = SVR(kernel='rbf') 
svr.fit(X_train, y_train) 
```

```python
y_pred1 = svr.predict(X_test) 
y_p1=[]
for i in y_pred1:
    y_p1.append(round(i))
y_p1
```

```python
y_pred1
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred1) 
```

<h1>CLASSIFICATION</h1>

```python
from sklearn import neighbors
from sklearn.metrics import accuracy_score

k=[]
n_acc_best=0
best=0
a_score=[]

for n in range(60):
  model = neighbors.KNeighborsRegressor(n_neighbors = n+1)
  model.fit(X_train, y_train)
  y_pred2=model.predict(X_test) 

  y_p2=[]
  for l in y_pred2:
    y_p2.append(round(l))

  i=accuracy_score(y_test, y_p2)
  a_score.append(round(i,3))

  k.append(n+1)

  if (i>best):
    n_acc_best=n+1
    best=i

print(n_acc_best)
```

```python
import matplotlib.pyplot as plt
plt.plot(k, a_score)  # Plot the chart 
plt.show()
```
